```python
def get_values(log, target):
    regex = re.compile(fr'{target}[ ]*=[ ]*\[?(.*?)\]?(?:,|\n|$)')
    values = re.findall(regex, log)
    values = [float(i) if ' ' not in i else i for i in values]
    return values
```

```python
def mt_split(mtlog):
    mtlog = [i.split() for i in mtlog]
    #print('--debug--')
    #print(list(zip(*mtlog)))
    #print('--debug--')
    return [float(i[0]) for i in mtlog], [float(i[1]) for i in mtlog], [float(i[2]) for i in mtlog]
    #return list(zip(*mtlog))

def plot_metric(attr, title, legend_loc='best'):
    stcnn_t1_val = get_values(stcnn_t1, attr)
    stcnn_t2_val = get_values(stcnn_t2, attr)
    stcnn_t3_val = get_values(stcnn_t3, attr)
    
    if attr == 'loss':
        attr = 'loss all'
    mtcnn_acc = get_values(mtcnn, attr)
    
    mtcnn_t1_val, mtcnn_t2_val, mtcnn_t3_val = mt_split(mtcnn_acc)
        
    epoch = list(range(33))
    plt.plot(epoch, stcnn_t1_val, '--r', label='single-task 1 toxicity')
    plt.plot(epoch, stcnn_t2_val, '--g', label='single-task 2 aggression')
    plt.plot(epoch, stcnn_t3_val, '--b', label='single-task 3 attack')
    plt.plot(epoch, mtcnn_t1_val, '-r', label='multi-task 1 toxicity')
    plt.plot(epoch, mtcnn_t2_val, '-g', label='multi-task 2 aggression')
    plt.plot(epoch, mtcnn_t3_val, '-b', label='multi-task 3 attack')

    plt.legend(loc=legend_loc)
    plt.xlabel('epoch')
    plt.ylabel('AUC')
    #plt.title(title)
    plt.xlim(0.0, 32.0)
    plt.show()
```
